```python
import tushare as ts
print ts.__version__
import pandas as pd
```

```python
df = ts.get_tick_data('600848', date='2018-12-12', src='tt')
```

```python
df.head(10)
```

```python
df['time'] = '2018-12-12 ' + df['time']
df['time'] = pd.to_datetime(df['time'])
```

```python
df = df.set_index('time')
```

```python
df.tail(10)
```

```python
price_df = df['price'].resample('1min').ohlc()
```

```python
price_df = price_df.dropna() #删除空值
```

```python
price_df.head(10)
```

```python
vols = df['volume'].resample('1min').sum()
```

```python
vols = vols.dropna()
print vols.head(5)
```

```python
vol_df = pd.DataFrame(vols, columns=['volume'])  # series转化为dataframe形式
```

```python
amounts = df['amount'].resample('1min').sum()
```

```python
amounts = amounts.dropna()
```

```python
amount_df = pd.DataFrame(amounts, columns=['amount'])
```

```python
newdf = price_df.merge(vol_df, left_index=True, 
                       right_index=True).merge(amount_df, 
                                               left_index=True, 
                                               right_index=True)
```

```python
newdf.tail(20)
```

```python
d_dict = {'open':'first',
          'high':'max',
          'close':'last',
          'low':'min',
          'volume':'sum', 
          'amount':'sum'}
```

```python
new = pd.DataFrame()
for col in newdf.columns:
    new[col] = newdf[col].resample('5min',how=d_dict[col])
```

```python
new.tail(20)
```
